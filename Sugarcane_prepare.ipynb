```python
import pandas as pd
import torch

trait_path = "E:/learning resource/PhD/phenotypes.csv"
genos_path = "E:/learning resource/PhD/geno_data1.csv"

phenotypes = pd.read_csv(trait_path,sep="\t").sample(2000)
genos = pd.read_csv(genos_path,sep="\t")
genos[genos=="--"] = 0.01
phenotypes[phenotypes == ""]
print(genos)
print(phenotypes.shape)
samples = phenotypes["Clone"]
print(samples)
merged_data = None
```

Merge geno data and phenotype data in to one dataframe;

```python
#genos.index = pd.Index(genos.columns,name="sample")
arrays = genos.T
print(arrays)
arrays.dropna(axis=0,how="any",inplace=True)
arrays.index = arrays.index.set_names("sample")
arrays = arrays.reset_index()

arrays.head(10)
```

```python
merged_data = pd.merge(phenotypes,arrays,left_on="Clone",right_on="sample")
merged_data.shape
merged_data.head(10)
#merged_data.to_csv("../merged_sugarcane.csv",sep="\t")
```

Specify X and Y;

```python
merged_data["Series"].unique()
merged_data["Region"].unique()
merged_data["Crop"].unique()
size = 2000
merged_data_small = merged_data.sample(size)
print(merged_data_small.shape)
traits = ["CCSBlup","TCHBlup","FibreBlup"]
Ys = pd.DataFrame()
for trait in traits:
    Y = merged_data_small[trait]
    Ys = pd.concat([Ys,Y],axis=1)
```
